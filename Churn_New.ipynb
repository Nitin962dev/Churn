### Importing the necessary libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import joblib
```

### Lord The Dataset 

```python
df = pd.read_csv('Clean_Churan.csv')
```

```python
df.head()
```

### Splitting the data for input and output 

```python
x = df.drop(columns = 'Exited') #Spliting the inputs 
x.head()

y = df['Exited'] #Spliting the output 
y.head()
```

```python
x.head()
```

```python
num_col = x.columns
```

```python
x.drop(columns = ['RowNumber', 'CustomerId', 'Surname'],inplace=True)
```

```python
x.head()
```

### Splitting the data for the training and tsting 

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

### creating the pipline

```python
cat_cols = [
    "Geography",
    "Gender"
]
```

```python
num_cols = [
    "CreditScore",
    "Age",
    "Tenure",
    "Balance",
    "NumOfProducts",
    "HasCrCard",
    "IsActiveMember",
    "EstimatedSalary"
]
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(), cat_cols)
    ]
)
```

```python
pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("model", LogisticRegression(class_weight="balanced"))
])
```

```python
#Training the model 
pipeline.fit(X_train,Y_train)
```

```python
y_pred = pipeline.predict(X_train)
```

```python
accuracy = accuracy_score(y_pred,Y_train)
accuracy
```

```python
joblib.dump(accuracy,"acc_log.pkl")
```

### Support Vector Regressor (SVR)


```python
pipeline_SVC = Pipeline([
    ("preprocessing", preprocessor),
    ("model", SVC(
    kernel='rbf',
    C=1.0,
    gamma='scale',
    random_state=42
))
])
```

```python
pipeline_SVC.fit(X_train, Y_train)
```

```python
y_pred_svc = pipeline_SVC.predict(X_test)
```

```python
accuracy_svc_test = accuracy_score(Y_test,y_pred_svc)
accuracy_svc_test
```

### K-Nearest Neighbors Regressor

```python
pipeline_KNN = Pipeline([
    ("preprocessing", preprocessor),
    ("model", KNeighborsRegressor(
        n_neighbors=5,
        weights='uniform',
        metric='minkowski',
        p=2
    ))
])
```

```python
pipeline_KNN.fit(X_train, Y_train)
```

```python
y_pred_knn = pipeline_KNN.predict(X_test)
```

```python
Y_test = Y_test.astype(int)
y_pred_knn = y_pred_knn.astype(int)

accuracy_knn_test = accuracy_score(Y_test, y_pred_knn)
accuracy_knn_test
```

```python
joblib.dump(accuracy_knn_test,"acc_knn.pkl")
```

```python
joblib.dump(pipeline_KNN,"churn_pipeline_KNN.pkl")
```
